```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from pycaret.regression import *
import pandas as pd
import pickle

# Step 1: Combine features and target
df_model = pd.concat([X, y], axis=1)
df_model.rename(columns={'runs_x': 'target'}, inplace=True)

# Step 2: PyCaret Setup
setup(data=df_model,
      target='target',
      session_id=123,
      use_gpu=False,  # Set False if no GPU
      normalize=True,
      transformation=True,
      remove_multicollinearity=True,
      multicollinearity_threshold=0.9,
      verbose=False)

# Step 3: Compare and select top 5 models
top_models = compare_models(n_select=5,
                            sort='RMSE',
                            exclude=['tr', 'catboost'],
                            fold=5)

# Step 4: Save models + their individual metrics
model_info_list = []

for i, model in enumerate(top_models):
    model_name = str(model).split("(")[0]
    filename = f"{model_name}_model_{i}"

    # Save model
    save_model(model, filename)
    model_file = filename + ".pkl"

    # Predict and pull individual metrics
    predict_model(model)
    model_metrics = pull()

    model_info_list.append({
        "name": model_name,
        "filename": model_file,
        "metrics": model_metrics.to_dict()
    })

# Step 5: Save all metadata
with open("cricket_model_info_list.pkl", "wb") as f:
    pickle.dump(model_info_list, f)

# Step 6: Blend top models
blended = blend_models(estimator_list=top_models, fold=5)
save_model(blended, 'blended_model')

print("✅ Models trained, individually evaluated, and saved!")
```
